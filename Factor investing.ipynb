```python
import pandas as pd
import statsmodels.api as sm

results = []

for stock in excess_return.columns:

    y = excess_return[stock]

    X = sm.add_constant(factors)

    model = sm.OLS(y, X).fit()

    results.append({
        'stock': stock,
        'alpha': model.params['const'],
        'mkt_beta': model.params['Mkt-RF'],
        'smb_beta': model.params['SMB'],
        'hml_beta': model.params['HML'],
        'r2': model.rsquared,
        'alpha_pvalue': model.pvalues['const']
    })

summary_df = pd.DataFrame(results)
```

```python
def max_drawdown(ret):
    cum = (1 + ret).cumprod()
    drawdown = cum / cum.cummax() - 1
    return drawdown.min()
```

```python
value_mdd = max_drawdown(value_portfolio)

growth_mdd = max_drawdown(growth_portfolio)

print(value_mdd)
print(growth_mdd)
```
